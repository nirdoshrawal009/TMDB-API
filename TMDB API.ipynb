# Fetch the trending TV Shows for the week from the TMDb API and print the taglines of the top 5 shows. If there is no tagline, print 'Empty' instead

```python
api_key = "5809a5a2864784e067048e640cbf1545"
params = {'api_key':api_key}
header = {'Accept': 'application/json'}
response = requests.get("https://api.themoviedb.org/3/trending/tv/week", params=params, headers = header)
data = response.json()
print(response.status_code)
print(response.url)
print(data)
```

```python
results=data.get('results')
ids=[]
for result in results[:5]:
    print(result.get('name'),end=' ')
    ids.append(result.get('id'))
    print(result.get('id'))
```

```python
for i in ids:
    api_key = "5809a5a2864784e067048e640cbf1545"
    params = {'api_key':api_key}
    header = {'Accept': 'application/json'}
    response = requests.get("https://api.themoviedb.org/3/tv/"+ str(i), params=params, headers = header)
    data = response.json()
    if data.get('tagline')!='':
        print(data.get('tagline'))
    else:
        print('Empty')
```
